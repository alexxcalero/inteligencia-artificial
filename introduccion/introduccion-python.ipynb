## <font color=blue>TIPOS BÁSICOS DE VARIABLES</font>
Las variables son "etiquetas" que permiten hacer referencia a los datos. Los nombres de variables válidos en Python deben ajustarse a las tres siguientes reglas:

* La variable puede ser una secuencia seguida de letras y dígitos.
* La secuencia debe empezar con una letra (no inicia nunca con un número).
* Además de a...z, y A...Z, el guión bajo (_) es aceptado.

### Variables numéricas

**Enteros**: Son aquellos números positivos o negativos que no tienen decimales. Se representan mediante el tipo ***int*** (abreviatura de entero).

```python
entero = 10
```

**Reales**: Los números reales son los que tienen decimales. En Python se expresan mediante el tipo ***float***.

```python
real = 0.3403
print(real)
type(real)
```

**Números Complejos**: Son una extensión de los números reales pero contienen la parte imaginaria. En Python se expresan mediante el tipo ***complex***.

```python
z = 2 + 3j
print(z)
type(z)
```

```python
y = 2 - 3j
```

```python
w = z + y
w
```

### Booleanos
Son representaciones a nivel binario. Una variable de tipo booleano solo puede tener dos valores: True (verdadero) y False (falso). Estos valores son muy importantes para las expresiones condicionales y los bucles. Los valores booleanos se representan mediante el tipo ***bool***.

```python
a = True
print(a)
type(a)
```

```python
a = 2
b = 3
c = 5
x = (b < c) | (b > a)
x
```

### Listas
Son colecciones ordenadas de items.
* Para acceder a los elementos se coloca el nombre de la lista seguido del índice del item en corchetes ([ ]). El primer item es el índice 0.
* La longitud de la lista de puede obtener con la función <b>len()</b>.
* Las listas son mutables (se puede cambiar los items).
* Las cadenas de caracteres son listas de caracteres inmutables.

```python
# Crea una lista llamada 'milista' con 4 items. El cuarto item es una cadena
milista = [1, 2, 3, "hola"]
print(milista)
```

```python
# Muestra el primer item de la lista milista
print(milista[0])

# Muestra el ultimo item de la lista milista
print(milista[-1])
```

```python
# Muestra la longitud (numero de items) de la lista milista
print(len(milista))
```

```python
# Imprime elemento por elemento de la lista
for item in milista:
    print(item)
```

```python
# Cambia el primer elemento de la lista
milista[0] = 3
print(milista)
```

```python
# Anhade un item al final de la lista
milista.append(33)
print(milista)
```

```python
# Crea una nueva lista (otralista) y la anexa como item a la lista milista
otralista = [7, 8, 9]
milista.append(otralista)
print(milista)
```

```python
# El primer corchete accede al ultimo elemento de la lista milista, y el segundo corchete accede al ultimo elemento de la sublista otralista
milista[-1][-1]
```

```python
# Extiende la lista milista con los elementos de una segunda lista (otralista)
otralista = [50, 60, 70]
milista.extend(otralista)
print(milista)
```

```python
# Borra el segundo elemento de milista
del milista[1]
milista
```

```python
# Lista de listas (matriz de 2 dimensiones)
a = [[77, 68, 86, 73], [96, 87, 89, 81], [70, 90, 86, 81]]
print(a[1][1])
```

### Tuplas
Son colecciones de items inmutables (de solo lectura). Se especifica los items separados por coma (paréntesis es opcional).

```python
a = (1, 2, 3)
print(a)
```

```python
type(a)
```

```python
a = (1,)  # Una tupla singleton con un entero
```

```python
type(a)
```

```python
b = ('hola',)  # Una tupla singleton con un string
```

El acceso a un item de la tupla se hace similar al caso de listas.

```python
a = (1, 2, 3)
a[1]  # Retorna el segundo elemento de la tupla
```

```python
# No se puede cambiar los items de una tupla (son inmutables)
a[1] = 10
```

```python
# Sin embargo, elementos de la tupla que son listas si podrian ser actualizados
x = (2, [5, 6], "hola")
print(x)
```

```python
# El segundo item de la tupla x es una lista, por lo tanto, se puede cambiar sus elementos
x[1].append(8)
x[1][1] = 7
x
```

Los elementos de la tupla se pueden <b>desempacar</b> en variables.

```python
tupla_estudiante = ('Maria', [15, 12, 16])
nombre, notas = tupla_estudiante
print(nombre)
print(notas)
```

```python
x = [[2, 3], 'Edwin', 'Villanueva']
notas, nombre, apellido = x
print(nombre)
print(apellido)
print(notas)
```

### Diccionarios
Son estructuras que almacenan entradas pares (<b>clave:valor</b>). Las claves (<b>keys</b>) deben ser valores inmutables (como strings y tuplas) y no duplicados. Los valores (<b>values</b>) pueden ser cualquer tipo de variable de Python.

```python
# Crea un diccionario con 3 entradas pares (clave:valor). Las claves son nombres y los valores edades
dic_edades = {'Maria': 25,
              'Jose' : 28,
              'Pedro': 35}
print(dic_edades)
```

```python
# Recupera el valor (edad) de la entrada cuyo key es 'Jose'
dic_edades['Jose']
```

```python
# Altera el valor del key 'Jose'
dic_edades['Jose'] = 30
print(dic_edades)
```

```python
# Recupera todas las claves del diccionario dic_edades
dic_edades.keys()
```

```python
# Recupera todos los valores del diccionario
dic_edades.values()
```

```python
# Recupera todos los items (pares key:value)
dic_edades.items()
```

### Sentencias condicionales
Condicionales <b>if ... else</b> ejecuta las líneas de código que están identadas debajo del if si la condición es True, caso contrario ejecuta las acciones que estan identadas debajo del else.

Condicionales <b>if ... elif ... else</b> ejecuta las acciones debajo del if o elif que cumpla la condición, caso contrario ejecuta las acciones debajo del else.

```python
# Pide un valor numerico al usuario y lo compara en los condicionales
valor = int(input('¿Cuál es tu edad?: '))

if 0 <= valor <= 20:
    print('Eres muy joven')
elif 20 < valor <= 40:
    print('Eres joven aún')
elif 40 < valor <= 60:
    print('Tu experiencia es valiosa')
else:
    print('Tienes mucho que enseñarnos')
```

```python
# Pide un string al usuario y lo compara en los condicionales
valor = str(input('¿Quién es el mejor jugador de la selección? '))
if valor == "Paolo Guerrero":
    print('Estoy de acuerdo')
else:
    print('No sabes nada')
```

### Bucle <b>While</b>
Repite la acción o grupo de acciones mientras se mantenga la condición.

```python
# Ejecuta las lineas debajo del while mientras i sea menor a 10
i = 0
while i < 10:
    print(i)
    i += 1  # esta sentencia equivale a i = i + 1
print('fin')
```

### Bucle <b>For</b>
Repite una acción o conjunto de acciones para cada item en la secuencia de items. Estas secuencias pueden ser listas o iterables construidos con la función <b>range()</b>

```python
# Recorre una lista con for e imprime sus items
milista = ["hola", "¿como estas?", "adios"]
for item in milista:
    print(item)
```

```python
# Recorre un diccionario con for e imprime sus items
for nombre, edad in dic_edades.items():
    print("Edad de {}: {}".format(nombre, edad))
```

```python
# Recorre un bucle for usando el iterador range()
for i in range(10):
    print("Valor de i =", str(i))
```

Note que range(i) genera una secuencia iterable de enteros entre 0 y i-1 (no incluye i). Se puede modificar el valor de inicio y los pasos, pasando como parametros a <b>range(inicio, maximo, pasos)</b>

```python
# Bucle for donde el iterador genera una secuencia iniciando en 5 hasta 25 y en pasos de 2
for i in range(5, 25, 2):
    print("valor de i =", str(i))
```

La sentencia <b>break</b> termina la ejecución del bucle for o while más interno.

```python
# Dos for anidados iterando de 0 a 4. Cuando la suma de los iteradores es 3 rompe el loop interno
for i in range(5):
    for j in range(5):
        if (i + j) == 3:
            break  # Rompe el for interno cuando i + j es igual a 3
        print ("i =", str(i), ", j =", str(j))
```

La sentencia <b>continue</b> termina la ejecución de la iteración actual en el bucle for o while más interno y continua con la siguiente iteración.

```python
# Itera de 0 a 7. Cuando el iterador es 4 se salta a la siguiente iteracion sin imprimir
for i in range(8):
    if i == 4:
        continue
    print(i)
```

## <font color=blue>FUNCIONES</font>

Una <b>función</b> es un fragmento de código con un nombre asociado que realiza una tarea específica y devuelve un valor. El programador puede especificar <b>argumentos</b> (valores que se le pasan a la función) para realizar la tarea. Si el programador no especifica un valor de retorno la función devuelve el valor None (nada).

```python
# Define una funcion que suma sus argumentos x, y
def suma(x, y):
    return x + y
```

```python
# Prueba la funccion suma que se acaba de definir
resultado = suma(1, 3)
print(resultado)
```

También, se puede pasar los nombres de los argumentos en cualquier orden.

```python
# Define una funcion que calcula el area de un rectangulo dados su largo y alto
def area_rectangulo(largo, alto):
    return largo * alto
```

```python
area_rectangulo(alto=5, largo=10)
```

También, se puede tener una lista de argumentos de tamaño arbitrario.

```python
def promedio(*args):
    return sum(args) / len(args)
```

```python
promedio(5, 10, 15, 20, 25)
```
