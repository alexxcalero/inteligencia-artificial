### Accediendo a datos de arrays numpy

Para acceder a una fila o columna entera de un array 2D se coloca entre corchetes el índice de la fila o columna que se quiere en la respectiva dimensión y la otra dimensión se indica con el simbolo <b>:</b> (significa todos los elementos de esa dimensión).

Por ejemplo, si X es un array de 2 dimensiones, entonces:
- X[ i , : ] es la fila i del array. Es equivalente a especificar X[ i , 0:len(X) ]
- X[ : , j ] es la columna j del array

Cuando se especifica una dimensión y todas las otras dimensiones de la derecha son :, entonces es opcional colocar los simbolos :

Por ejemplo:
- En un array 2D, X[ i, : ] es lo mismo que X[ i ] (fila i)
- En un array 3D, X[ i, : , : ] es lo mismo que X[ i ] (matriz i)

```python
x = [[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]]
print(x)
```

```python
# Crea un array 2D de 3x4
miarray = np.array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
print(miarray)
```

```python
# Para acceder al elemento de la fila 1, columna 2 del array
miarray[1, 2]
```

```python
# Obtiene la tercera fila del array
miarray[2, :]  # Esto es lo mismo que miarray[2, 0:4]
```

```python
# Otra forma mas simplificada de obtener la tercera fila del array
miarray[2]  # Igual a miarray[2, :]
```

```python
# Obtiene la columna 2 del array:
x = miarray[:, [2]]
x
```

```python
# Ultima fila:
x = miarray[-1, :]
x
```

#### Slicing
El obtener submatrices de arrays (slices) se llama **slicing**. Esto se hace indicando los índices de las filas y columnas (o dimensiones superiores) que se quiere recuperar. En caso que sean filas o columnas consecutivas se puede usar el operador :

Por ejemplo:
- Una indexación 1:3 en la dimensión de columnas significa "las columnas desde la columna 1 hasta la columna 3 (no incluye la columna 3)"
- Una indexación 0:3 en la dimensión de filas significa "las filas desde la fila 0 hasta la fila 3 (no incluye la fila 3)". Es opcional poner 0, es decir es lo mismo la indexación 0:3 o la idexacion :3

image.png

```python
miarray = np.array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
a_slice = miarray[0:2, 1:3]  # Obtiene un slice de filas 0, 1 y columnas 1, 2
a_slice
```

```python
a_slice[0, 1]  # Accede a la fila 0, columna 1 del slice que acaba de extraer
```

```python
miarray
```

```python
a_slice
```

```python
a_slice[0, 1] = -10
a_slice
```

```python
# Slice con todas las filas y columnas 1, 2, 3
miarray[:, 1:4]
```

```python
# Slice con filas 0, 2  y todas las columnas
y = miarray[[0, 2], :]
y
```

### Haciendo copias profundas de arrays
Para estar seguro de hacer una copia completa e independiente de un array se puede usar su metodo <b>.copy</b>

```python
# Define un array
miarray = np.array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
miarray
```

```python
# Obtiene un slice de filas 0, 1 y columnas 1, 2
a_slice = miarray[0:2, 1:3]
a_slice
```

```python
a_slice[0, 1] = 5  # Coloca el valor 5 en la posicion 0, 1 de a_slice
a_slice
```

```python
miarray
```

Vemos que la asignación de valor en la sentencia anterior toma efecto también en el array original. En realidad los slices no son copias independientes del array original, sino solo vistas de la misma. Para hacer copias independientes de un array debemos usar el metodo .copy()

```python
otroarray = miarray.copy()  # Hace una copia independiente de miarray en otra variable llamada otroarray
otroarray
```

```python
miarray[0, 0] = 100
print(miarray)
```

```python
print(otroarray)  # Vemos que no afecta la signacion anterior en otroarray
```

```python
otroarray[0,0] = 50  # Lo mismo si cambiamos otroarray, no afecta miarray
miarray
```

### Operaciones aritméticas con arrays
Operaciones con array son hechas a nivel de elementos (element-wise).

```python
# Eleva al cuadrado los elementos del array
x = np.array([1, 2, 3, 4, 5])
x = x ** 2
x
```

```python
# Suma 2 arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 11, 12, 13, 14])
z = x + y
z
```

```python
# Multiplicacion de arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([10, 11, 12, 13, 14])
z = x * y
z
```

```python
np.matmul(x, y)
```

```python
# Suma un escalar
z = x + 1
z
```

```python
# Comparacion
x = np.array([1, 2, 3, 4, 5])
mascara = x != 5
mascara
```

```python
x[(x > 1) & (x < 5)]
```

```python
(x > 1) & (x < 5)
```

```python
# Comparacion
x = np.array([1, 0, 3, 4, 0])
mascara = x != 0
mascara
```

```python
# Indexacion booleana
z = x[mascara].copy()  # Mascara es un array de booleanos de tamaño de x. Solo las posiciones de mascara en True seran devueltas
z
```

```python
# Devuelve los elementos de x que sean mayores a 0 y menores a 4
z = x[(x > 0) & (x < 4)]  # El & es el operador AND. El | es el OR
z
```

### Operaciones de <b>reducción</b> con arrays Numpy

Operaciones de reducción disminuyen la información del array a arrays de menos dimensiones o incluso números escalares.

```python
x = np.array([[87, 96, 70], [100, 87, 90], [94, 77, 90], [100, 81, 82]])
x
```

```python
# Suma total de valores
x.sum()
```

```python
# Minimo valor
x.min()
```

```python
# Maximo valor
x.max()
```

```python
# Media
x.mean()
```

```python
# Desviacion estandar
x.std()
```

Se puede hacer también las operaciones de reducción a nivel de filas o columnas especificando el argumento <b>axis</b> (axis=0: medias de las columnas, axis=1: medias de las filas)

```python
x
```

```python
x.shape
```

```python
# Reduce el axis 0 (filas) con la media, osea calcula los valores medios de las columnas
y = x.mean(axis=0)
y
```

```python
y.shape
```

```python
# Calcula los valores medios de las filas
x.mean(axis=1)
```
